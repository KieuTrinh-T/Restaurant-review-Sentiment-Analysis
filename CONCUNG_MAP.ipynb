```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
```

```python
#internal data
df_ext = pd.read_csv('concung.csv')
#internal data
df_int = pd.read_excel('CONCUNG.xlsx',
                     sheet_name='CONCUNG BI') 
```

# PREPROCESSING

```python
df_int.info()
```

```python
df_int.columns
```

```python
df_int.drop(columns=['Unnamed: 43', 'Unnamed: 44'], inplace=True)
```

```python
df_int[df_int['LATITUDE'].isnull()].info()
```

```python
df_int.dropna(subset='LATITUDE', axis=0, inplace=True)
```

```python
df_int['PROVINCE_NAME_SHORT'] = df_int['PROVINCE_NAME'].str.split(' ',1).str[-1]
```

```python
df_int['PROVINCE_NAME_SHORT']
```

```python
df_int.info()
```

# JOIN 2 SOURCES

```python
df_merged = pd.merge(df_ext, df_int, how='right', left_on=['province'], right_on=['PROVINCE_NAME_SHORT'])
df_merged[['CUSTOMER_NAME','ADDRESS_x']].head()
```

```python
df_merged
```

```python
df_merged['Distance'] = df_merged.apply(lambda x: geodesic((x['LATITUDE_x'], x['LONGITUDE_x']), (x['LATITUDE_y'], x['LONGITUDE_y'])).m if pd.notnull(x['LATITUDE_x']) else np.nan, axis=1)
```

```python
df_merged[['CUSTOMER_CODE','ADDRESS_y','ADDRESS_x', 'Distance']].where(df_merged['Distance'] < 150).sort_values(by=['Distance'], ascending=False).dropna()
```

```python
#Get matched
df_match = df_merged[['CUSTOMER_CODE','ADDRESS_y', 'STORE_CODE','ADDRESS_x']].where(df_merged['Distance'] < 150).rename(columns={
                            'ADDRESS_y':'ADDRESS_INTERNAL',
                            'ADDRESS_x':'ADDRESS_EXTERNAL'
})
```

```python
df_match.to_csv('CONCUNG_MATCHED.csv')
```

```python
  
#Resolve non match customer code
df_non_m = df_merged[~df_merged['CUSTOMER_CODE'].isin(df_match['CUSTOMER_CODE'])]
df_non_m
```

```python
df_non_m[['CUSTOMER_CODE','ADDRESS_y', 'STORE_CODE','ADDRESS_x']]
```

```python
df_non_m[['STREET_NAME']].value_counts()
```

```python
df_non_m['MATCHED_STREET'] = df_non_m.apply(lambda x: True if x['ADDRESS_x'].find(str(x['STREET_NAME'])) != -1 else False, axis=1)
df_non_m[['CUSTOMER_CODE','ADDRESS_y', 'STORE_CODE','ADDRESS_x', 'MATCHED_STREET']]
```

```python
df_non_m[['CUSTOMER_CODE','ADDRESS_y', 'STORE_CODE','ADDRESS_x','Distance', 'MATCHED_STREET']].where(df_non_m['MATCHED_STREET']==True).dropna(subset=['CUSTOMER_CODE'], axis=0)
```

```python
df_non_m['CUSTOMER_CODE'].value_counts()
```

```python
df_filtered=df_non_m[['CUSTOMER_CODE','ADDRESS_y', 'STORE_CODE','ADDRESS_x','Distance', 'MATCHED_STREET']] \
        .where((df_non_m['ward']==df_non_m['WARD_NAME']) & (df_non_m['MATCHED_STREET']==True))\
        .dropna(subset='CUSTOMER_CODE', axis=0) \
        .sort_values(by=['Distance'], ascending=False )
```

```python
df_non_m[~df_non_m['CUSTOMER_CODE'].isin(df_filtered['CUSTOMER_CODE'])][['CUSTOMER_CODE','ADDRESS_y', 'STORE_CODE','ADDRESS_x']]
```

```python
df_match_2 = pd.concat([df_match, df_filtered[['CUSTOMER_CODE','ADDRESS_y', 'STORE_CODE','ADDRESS_x']]\
                .rename(columns={
                            'ADDRESS_y':'ADDRESS_INTERNAL',
                            'ADDRESS_x':'ADDRESS_EXTERNAL'
})], axis=0)
df_match_2
```

```python
df_match_2.dropna(axis=0)
```

```python
df_match_2.to_csv('CONCUNG_MATCHED.csv')
```

```python
df_non_m=df_non_m[~df_non_m['CUSTOMER_CODE'].isin(df_filtered['CUSTOMER_CODE'])][['CUSTOMER_CODE','ADDRESS_y', 'STORE_CODE','ADDRESS_x']]
```
